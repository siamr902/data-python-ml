**Matplotlib Library**
- Used in creating plots to visualize data

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(f"{x}\n")
print(f"{y}\n")
print(z)
```

**Plotting the data**

```python
plt.plot(x, y)
plt.show()
```

```python
# cosine wave
plt.plot(x, z)
plt.show()
```

```python
# adding title, x-axis and y-axis labels
```

```python
plt.plot(x, y)
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("sine wave")
plt.show()
```

**Parabola and Cubic Functions**

```python
x = np.linspace(-10, 10, 20)
y = pow(x, 2)
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-10, 10, 20)
y = pow(x, 3)
plt.plot(x, y)
plt.show()
```

**Colors**

```python
x = np.linspace(0, 80, 80)
y = pow(np.e, x)
plt.plot(x, y, "r+")
plt.show()
```

```python
# green dots
plt.plot(x, y, "g.")
plt.show()
```

```python
plt.plot(x, y, "bx")
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), "g-") # line
plt.plot(x, np.cos(x),"r--") # dotted
```

**Bar Plots**

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
languages = ["English", "German", "Spanish", "Mandarin", "Latin"]
people = [100, 50, 200, 40, 70]
ax.bar(languages, people)
plt.xlabel("LANGUAGES")
plt.ylabel("NUMBER OF PEOPLE")
plt.show()
```

**Pie Chart**

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
languages = ["English", "German", "Spanish", "Mandarin", "Latin"]
people = [100, 50, 200, 40, 70]
ax.pie(people, labels=languages, autopct="%1.1f%%")
plt.show()
```

**Scatter Plot**

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color="b")
ax.scatter(x, z, color="r")
plt.show()
```

**3D Scatter Plot**


```python
figure3 = plt.figure()
ax = plt.axes(projection="3d")
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap="Blues")
plt.show()
```
